```python
from nsepython import equity_history
from pathlib import Path
import plotly.graph_objects as go
import pandas as pd
from datetime import datetime, timedelta

def generate_dates(x):
    # Get today's date
    today = datetime.today().date()

    # Calculate X months ago
    # Create a timedelta object for X months
    delta = timedelta(days=30 * x)  # Approximating a month as 30 days

    # Subtract delta from today's date to get X months ago
    months_ago = today - delta

    # Format dates as strings in dd-mm-yyyy format
    today_str = today.strftime('%d-%m-%Y')
    months_ago_str = months_ago.strftime('%d-%m-%Y')

    return today_str, months_ago_str

ROOT_DIR = Path(r"C:\Users\anura\OneDrive\Documents\Finances\portfolio")
latest_dump = "2024-10"
LATEST_DIR = Path(ROOT_DIR, latest_dump)

portfolio_df = pd.read_excel(Path(LATEST_DIR, "Equity_Summary_Details.xls"), skiprows=[0,1,2,3])
equity_map = pd.read_csv(Path(ROOT_DIR, "equity_list_nse.csv"))
equity_map.columns = [x.strip() for x in equity_map.columns]

portfolio_df = portfolio_df.rename(columns = {"ISIN":"ISIN NUMBER"})
merged_df = portfolio_df.merge(equity_map, on="ISIN NUMBER", how="left")
merged_df = merged_df.dropna()
merged_df.columns = [x.lower() for x in merged_df.columns]
merged_df.head()
```

```python
purchase_df = pd.read_excel(Path(LATEST_DIR, "Your Recent Purchase 6677164.xls"), skiprows=[0,1,2,3]).dropna()
purchase_df.columns = [x.lower() for x in purchase_df.columns]
purchase_df
```

```python
data_dict = {}
end_date, start_date = generate_dates(6)
start_date = "01-01-2024"

for symbol in merged_df.SYMBOL:
    print(symbol)
    series = "EQ"
    temp_df = equity_history(symbol,series,start_date,end_date)
    temp_df = temp_df.sort_values(by = "CH_TIMESTAMP").reset_index(drop=1)
    data_dict[symbol] = temp_df
    break
```

```python
symbol = "BAJFINANCE"

df = data_dict[symbol]

# Create a figure
fig = go.Figure()

# Add candlestick trace
fig.add_trace(go.Candlestick(x=df['CH_TIMESTAMP'],
                             open=df['CH_OPENING_PRICE'],
                             high=df['CH_TRADE_HIGH_PRICE'],
                             low=df['CH_TRADE_LOW_PRICE'],
                             close=df['CH_CLOSING_PRICE'],
                             name='Candlestick'))

# Customize layout
fig.update_layout(
    title=f'Candlestick Chart: {symbol}',
    xaxis_title='Timestamp',
    yaxis_title='Price',
    template='plotly_white',  # Use a light theme
    showlegend=True,  # Show legend
    hovermode='x',  # Show closest data on hover
    plot_bgcolor='rgba(0,0,0,0)'  # Transparent plot background
)

# Show the plot
fig.show()
```
